# How do you choose the right visualization for your data?

## Explanation {-}

The best plot depends on your **data type** and the **story you want to tell**. Here's a quick guide:

| Goal                              | Recommended Plots                             |
|----------------------------------|-----------------------------------------------|
| Show distribution                | Histogram, Density Plot, Boxplot, Violin Plot |
| Compare categories or groups     | Bar Plot, Dot Plot, Swarm Plot, Boxplot       |
| Explore relationships            | Scatter Plot, Line Plot, Pair Plot            |
| Visualize proportions            | Pie Chart, Donut Chart, Treemap               |
| Show set overlaps                | Venn Diagram, UpSet Plot                      |
| Display trends or time series    | Line Plot, Area Chart                         |

- Favor bar plots over pie charts for comparison clarity  
- Use hue/grouping to show categorical differences
